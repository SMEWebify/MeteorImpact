```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display, clear_output
import plotly.figure_factory as ff
from scipy.cluster.hierarchy import linkage

path = r"C:\Users\alien\Downloads\meteorite-landings.csv"

def pick_col(df, candidates):
    lower_map = {c.lower(): c for c in df.columns}
    for cand in candidates:
        if cand.lower() in lower_map:
            return lower_map[cand.lower()]
    return None

def to_num(s):
    return pd.to_numeric(
        s.astype(str).str.replace(",", ".", regex=False).str.strip(),
        errors="coerce",
    )

# -------------------------
# LOAD + CLEAN
# -------------------------
try:
    df_raw = pd.read_csv(path, sep=";")

    col_year = pick_col(df_raw, ["year_num", "year", "Year / AnnÃ©e", "AnnÃ©e", "annee", "Year"])
    col_country = pick_col(df_raw, ["country", "Country / Pays", "Pays", "Country"])
    col_mass = pick_col(df_raw, ["mass_g", "Mass (g) / Masse (g)", "Masse (g)", "Mass (g)", "mass", "mass_g"])
    col_lat = pick_col(df_raw, ["lat", "latitude", "Latitude"])
    col_lon = pick_col(df_raw, ["lon", "longitude", "Longitude"])

    missing = [name for name, col in [
        ("annÃ©e", col_year),
        ("pays", col_country),
        ("masse", col_mass),
        ("latitude", col_lat),
        ("longitude", col_lon),
    ] if col is None]

    if missing:
        raise ValueError(
            "Colonnes introuvables: " + ", ".join(missing) +
            "\nColonnes disponibles: " + ", ".join(df_raw.columns)
        )

    df = df_raw.rename(columns={
        col_year: "year_num",
        col_country: "country",
        col_mass: "mass_g",
        col_lat: "lat",
        col_lon: "lon",
    }).copy()

    df["mass_g"] = to_num(df["mass_g"])
    df["lat"] = to_num(df["lat"])
    df["lon"] = to_num(df["lon"])

    y = pd.to_numeric(df["year_num"], errors="coerce")
    if y.isna().mean() > 0.5:
        dt = pd.to_datetime(df["year_num"], errors="coerce")
        df["year_num"] = dt.dt.year
    else:
        df["year_num"] = y

    df = df.dropna(subset=["mass_g", "year_num", "lat", "lon", "country"]).copy()
    df["year_num"] = df["year_num"].astype(int)
    df["country"] = df["country"].astype(str).str.strip()

    df = df[
        (df["mass_g"] > 0) &
        (df["year_num"] > 0) &
        (df["lat"].between(-90, 90)) &
        (df["lon"].between(-180, 180))
    ].copy()

    df = df[~((df["lat"] == 0) & (df["lon"] == 0))].copy()

    print(f"Dataset chargÃ© : {len(df):,} lignes")
    print("Colonnes utilisÃ©es :", {"year_num": col_year, "country": col_country, "mass_g": col_mass, "lat": col_lat, "lon": col_lon})

except Exception as e:
    df = None
    print(f"âŒ Erreur : {e}")


# -------------------------
# DASHBOARD
# -------------------------
def update_dashboard(mass_min, year_range, top_n, top_mode, selected_countries, show_dendro, show_treemap):
    if df is None or df.empty:
        print("âŒ Dataset non chargÃ©.")
        return

    y0, y1 = year_range
    if y0 > y1:
        y0, y1 = y1, y0

    dff = df[
        (df["mass_g"] >= mass_min) &
        (df["year_num"] >= y0) &
        (df["year_num"] <= y1)
    ].copy()

    # Filtre pays (optionnel)
    if selected_countries:
        dff = dff[dff["country"].isin(selected_countries)].copy()

    if dff.empty:
        print("Aucune donnÃ©e trouvÃ©e avec ces filtres.")
        return

    # KPI
    print(
        f"ðŸ“Œ Points: {len(dff):,} | Pays: {dff['country'].nunique():,} | "
        f"AnnÃ©es: {dff['year_num'].min()}â€“{dff['year_num'].max()} | "
        f"Masse max: {dff['mass_g'].max():,.0f} g"
    )

    # Ranking Top N (count ou masse)
    if top_mode == "mass":
        rank = dff.groupby("country")["mass_g"].sum().sort_values(ascending=False)
    else:
        rank = dff["country"].value_counts()

    top_countries = rank.head(top_n).index
    dff_top = dff[dff["country"].isin(top_countries)].copy()

    # G1 Bar: (toujours count, câ€™est lisible)
    top_df = dff_top["country"].value_counts().reset_index()
    top_df.columns = ["country", "count"]
    fig1 = px.bar(
        top_df, x="count", y="country",
        orientation="h",
        title=f"Top {top_n} des pays (nombre d'impacts)",
        template="plotly_dark",
        color="count"
    )
    fig1.update_layout(yaxis={"categoryorder": "total ascending"}, height=360)

    # G2 Pie: somme des masses (Top N + Autres, calculÃ© sur dff complet filtrÃ©)
    mass_by_country = (
        dff.groupby("country", as_index=False)["mass_g"]
        .sum()
        .sort_values("mass_g", ascending=False)
    )

    pie_df = mass_by_country.head(top_n).copy()
    pie_df.columns = ["country", "mass_sum"]

    others_mass = mass_by_country.iloc[top_n:]["mass_g"].sum()
    if others_mass > 0:
        pie_df = pd.concat(
            [pie_df, pd.DataFrame([{"country": "Autres", "mass_sum": others_mass}])],
            ignore_index=True
        )

    fig2 = px.pie(
        pie_df,
        names="country",
        values="mass_sum",
        title=f"RÃ©partition de la masse totale (Top {top_n} + Autres)",
        template="plotly_dark"
    )
    fig2.update_layout(height=360)

    figs = [fig1, fig2]

    # G3 Dendrogram (optionnel)
    if show_dendro and len(dff_top["country"].unique()) >= 2:
        agg = (
            dff_top.groupby("country")["mass_g"]
            .agg(count="count", mean="mean", median="median", max="max", std="std")
            .fillna(0)
            .reset_index()
        )

        X = agg[["count", "mean", "median", "max", "std"]].to_numpy(dtype=float)
        X = (X - X.mean(axis=0)) / (X.std(axis=0) + 1e-9)
        labels = agg["country"].tolist()

        fig3 = ff.create_dendrogram(X, labels=labels, linkagefun=lambda x: linkage(x, method="ward"))
        fig3.update_layout(
            template="plotly_dark",
            title="Dendrogramme â€” SimilaritÃ© des pays (stats de masse + volume)",
            height=520,
            margin={"r":20,"t":60,"l":20,"b":20}
        )
        figs.append(fig3)

    # G4 Treemap (optionnel)
    if show_treemap:
        treedf = (
            dff_top.groupby("country", as_index=False)
            .size()
            .rename(columns={"size": "count"})
        )

        fig4 = px.treemap(
            treedf,
            path=["country"],
            values="count",
            color="count",
            title="Treemap â€“ RÃ©partition des impacts par pays",
            template="plotly_dark"
        )
        fig4.update_layout(height=520, margin={"r":0,"t":50,"l":0,"b":0})
        figs.append(fig4)

    display(*figs)


# -------------------------
# UI
# -------------------------
if df is not None and not df.empty:
    mass_slider = widgets.FloatLogSlider(
        value=1000, base=10, min=0, max=7, step=0.1,
        description="Masse min (g)",
        continuous_update=False,
        readout_format=".0f"
    )

    year_slider = widgets.IntRangeSlider(
        value=[max(1600, int(df["year_num"].min())), int(df["year_num"].max())],
        min=int(df["year_num"].min()),
        max=int(df["year_num"].max()),
        step=1,
        description="AnnÃ©es",
        continuous_update=False
    )

    top_slider = widgets.IntSlider(
        value=10, min=5, max=30, step=1,
        description="Nb pays",
        continuous_update=False
    )

    top_mode_widget = widgets.Dropdown(
        options=[("Nombre d'impacts", "count"), ("Masse totale", "mass")],
        value="count",
        description="Top basÃ© sur",
        style={"description_width": "initial"}
    )

    country_filter = widgets.SelectMultiple(
        options=sorted(df["country"].unique()),
        value=(),
        description="Pays (optionnel)",
        rows=8,
        style={"description_width": "initial"}
    )

    show_dendro = widgets.Checkbox(value=True, description="Afficher dendrogramme")
    show_treemap = widgets.Checkbox(value=True, description="Afficher treemap")

    out = widgets.Output()

    def on_change(_=None):
        with out:
            clear_output(wait=True)
            update_dashboard(
                mass_slider.value,
                year_slider.value,
                top_slider.value,
                top_mode_widget.value,
                list(country_filter.value),
                show_dendro.value,
                show_treemap.value
            )

    for w in [mass_slider, year_slider, top_slider, top_mode_widget, country_filter, show_dendro, show_treemap]:
        w.observe(on_change, names="value")


    ui = widgets.VBox([
        widgets.HBox([mass_slider, top_slider, top_mode_widget]),
        year_slider,
        widgets.HBox([show_dendro, show_treemap]),
        country_filter
    ])
```

```python
display(ui, out)

on_change()
```
